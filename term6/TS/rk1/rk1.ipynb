## 1. Выбранная открытая техническая система

В качестве примера открытой технической системы выбрана **CI/CD-система программного проекта**, построенная на базе GitHub Actions.

Данная система относится к области программной инженерии и является **открытой**, поскольку постоянно обменивается информацией с внешней средой: получает изменения исходного кода от разработчиков, обращается к внешним репозиториям зависимостей, разворачивает приложение в тестовой или рабочей инфраструктуре, отправляет уведомления и предоставляет результаты своей работы пользователям и команде разработки.

## 2. Надсистема и состав системы

### Надсистема (среда)

Надсистемой для CI/CD-системы является **среда разработки и эксплуатации программного продукта**, в которую входят:

- команда разработчиков;
- система хранения исходного кода (GitHub-репозиторий);
- тестовая и производственная инфраструктура;
- внешние пакетные реестры и сервисы зависимостей;
- системы мониторинга, логирования и уведомлений;
- заказчик, пользователи и бизнес-процессы, ради которых развивается программный продукт.

### Основные подсистемы и элементы

| Подсистема | Основные элементы | Назначение |
|---|---|---|
| Подсистема источников изменений | разработчики, git-коммиты, pull request, файлы workflow | Формирование событий, запускающих процесс CI/CD |
| Подсистема управления процессом | GitHub Actions, триггеры, оркестратор workflow | Определение сценария выполнения и координация этапов |
| Исполнительная подсистема | runners, контейнеры, виртуальные окружения | Непосредственное выполнение сборки, тестов и проверок |
| Подсистема контроля качества | unit/integration tests, линтеры, статический анализ, security scan | Проверка качества и корректности программного продукта |
| Подсистема артефактов | пакеты, docker-образы, кеши, хранилище артефактов | Сохранение результатов сборки и подготовка к развертыванию |
| Подсистема развертывания | staging, production, deployment scripts | Доставка новой версии в целевую среду |
| Подсистема обратной связи | логи, метрики, статусы, уведомления в почту/мессенджеры | Информирование участников и контроль результата |


## 3. Структура системы и структурная схема

Структура системы представляет собой совокупность **информационных, управляющих и обратных связей** между подсистемами. В системе можно выделить иерархический уровень управления (оркестратор workflow), исполнительный уровень (runners), а также уровень контроля и обратной связи (тесты, логи, уведомления).

Основные связи в системе:

- разработчик передает изменения в репозиторий;
- репозиторий формирует событие запуска pipeline;
- оркестратор направляет команды runner-ам;
- runner-ы выполняют сборку, тесты и проверки;
- результаты передаются в хранилище артефактов и в подсистему развертывания;
- статусы, логи и метрики возвращаются команде разработки как обратная связь.

### Структурная схема

```text
[Разработчик]
      |
      v
[Git-репозиторий] --> [Событие push / pull request] --> [Оркестратор CI/CD]
                                                            |
                                                            v
                                                     [Runner / Агент]
                                                            |
                ---------------------------------------------------------------
                |                         |                    |                |
                v                         v                    v                v
            [Сборка]                  [Тесты]        [Анализ кода]   [Security Scan]
                \_________________________|____________________|________________/
                                          |
                                          v
                                   [Артефакты сборки]
                                          |
                                          v
                              [Подсистема развертывания]
                                   |                |
                                   v                v
                           [Staging-среда]    [Production-среда]
                                   \                /
                                    \              /
                                     v            v
                              [Логи, метрики, уведомления]
```

## 4. Входы, выходы и возможные состояния системы

### Входы системы

- изменения исходного кода;
- pull request и merge request;
- конфигурационные файлы pipeline;
- секреты и переменные окружения;
- версии внешних библиотек и зависимостей;
- ручные команды запуска;
- текущее состояние серверов и целевых сред.

### Выходы системы

- собранные артефакты (пакеты, docker-образы, бинарные файлы);
- отчеты о тестировании;
- результаты статического анализа и security scan;
- развернутая версия приложения;
- журналы выполнения, метрики и статусы pipeline;
- уведомления разработчикам и администраторам.

### Возможные состояния системы

| Состояние | Содержание состояния |
|---|---|
| Ожидание | Система готова к приему нового события |
| Инициализация | Определяется workflow, резервируются ресурсы, подготавливается runner |
| Выполнение | Идут сборка, тесты, анализ, упаковка и иные шаги pipeline |
| Успешное завершение | Все этапы пройдены, артефакт готов или версия развернута |
| Ошибка | Один или несколько этапов завершились неуспешно |
| Приостановка | Pipeline остановлен вручную или заблокирован политиками |
| Деградация | Часть функций недоступна, например из-за отказа runner-а или внешнего реестра |


## 5. Цель функционирования системы и критерии эффективности

### Цель функционирования

Основная цель CI/CD-системы состоит в том, чтобы **автоматизированно, быстро и надежно превращать изменения в исходном коде в проверенный и готовый к поставке программный продукт**.

### Критерии эффективности

| Критерий | Смысл |
|---|---|
| Время выполнения pipeline | Чем меньше время от коммита до результата, тем выше оперативность системы |
| Процент успешных прогонов | Характеризует стабильность и надежность процессов |
| Частота поставки изменений | Показывает способность системы поддерживать непрерывную разработку |
| Доля дефектов, дошедших до production | Чем меньше значение, тем эффективнее контроль качества |
| Среднее время восстановления после сбоя | Показывает адаптивность и управляемость системы |
| Затраты вычислительных ресурсов | Характеризуют экономичность функционирования |
| Полнота обратной связи | Определяет, насколько быстро разработчик получает полезную информацию о результате изменений |


## 6. Дополнительная классификация системы

Кроме того, рассматриваемая система может принадлежать к следующим классам систем.

| Классификационный признак | Класс системы | Обоснование |
|---|---|---|
| По происхождению | искусственная, техническая | Система создана человеком и реализована на программно-технических средствах |
| По взаимодействию со средой | открытая | Непрерывно обменивается информацией и командами с внешней средой |
| По изменению во времени | динамическая | Состояние системы меняется при поступлении событий и в ходе выполнения pipeline |
| По характеру процессов | дискретная | Работа описывается последовательностью дискретных событий и состояний |
| По степени определенности | вероятностная (стохастическая) | Возможны случайные сбои, задержки сети, нестабильность внешних сервисов |
| По структуре | иерархическая, распределенная | Есть уровень управления и множество распределенных исполнительных узлов |
| По наличию управления | управляемая, кибернетическая | Система получает управляющие воздействия, контролирует результат и использует обратную связь |
| По целевой направленности | целеориентированная | Все подсистемы работают ради выпуска качественного программного продукта |
| По сложности | сложная | Содержит множество взаимосвязанных компонентов, состояний и внешних связей |


## 7. Признаки, доказывающие, что объект является системой

CI/CD-система действительно является системой, поскольку обладает следующими основными признаками:

1. **Наличие элементов.** Система состоит из множества компонентов: репозитория, оркестратора, runner-ов, тестов, хранилища артефактов, средств развертывания и мониторинга.
2. **Наличие структуры.** Между компонентами существуют устойчивые связи и определенный порядок взаимодействия.
3. **Целостность.** Компоненты работают не изолированно, а как единое целое, направленное на выпуск новой версии программного продукта.
4. **Границы системы.** Можно отделить внутренние компоненты CI/CD от внешней среды: разработчиков, облачной инфраструктуры, пользователей, внешних реестров.
5. **Входы и выходы.** Система принимает входные воздействия и выдает результаты своей работы во внешнюю среду.
6. **Управление и обратная связь.** Результаты тестов, статического анализа и развертывания используются для корректировки дальнейших действий.
7. **Цель функционирования.** Работа системы подчинена конкретной цели: автоматизировать интеграцию, проверку и поставку программного продукта.
8. **Эмерджентность.** Полезный результат возникает только при совместной работе всех подсистем; по отдельности ни runner, ни репозиторий, ни тесты не обеспечивают полный цикл поставки.

Следовательно, выбранный объект удовлетворяет основным системным признакам и может рассматриваться как полноценная открытая техническая система.
